```python
# -----------------------------
# Simple NLP Chatbot using CountVectorizer (no tokenization function)
# -----------------------------

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# -----------------------------
# Knowledge Base
# -----------------------------
answers = [
    "hello! i am your chatbot.",
    "i can help you with python, machine learning, and nlp basics.",
    "python is a programming language.",
    "machine learning is teaching computers to learn patterns.",
    "nlp stands for natural language processing.",
    "goodbye and have a nice day!"
    #Do you teach python" [-1] # do you train in pyton # cani learn python here
]

# -----------------------------
# Chatbot Response Function
# -----------------------------
def chatbot_response(user_input):
    user_input = user_input.lower()
    all_sentences = answers + [user_input]
    print(type(all_sentences))

    # Vectorize using CountVectorizer
    vectorizer = CountVectorizer()
    vectors = vectorizer.fit_transform(all_sentences)
    print(vectors.toarray())

    # Compute cosine similarity
    similarity = cosine_similarity(vectors[-1], vectors[:-1])
    print("Smilarity" ,similarity)

    # finding the Similarity Score
    index = similarity.argsort()[0][-1]  # best match index # sorting the similarity Score
    print("index with highest score or match" ,index) # 1
    score = similarity[0][index] # similarity[0][1]
    print("Score",score)


    if score > 0.2:
        return answers[index]
    else:
        return "I am sorry, I donâ€™t understand that."

# -----------------------------
# Chatbot Loop
# -----------------------------
print("Chatbot: Hello! Type 'bye' to exit.")

while True:
    user_input = input("You: ")
    if user_input.lower() in ["bye", "exit", "quit"]:
        print("Chatbot: Goodbye! ðŸ‘‹")
        break
    else:
        print("Chatbot:", chatbot_response(user_input))
```
